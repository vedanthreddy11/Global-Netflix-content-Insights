**CONVERTING OBJECT DATATYPE TO STRING (date_added,duration col contains both numeric and categorical values)**

```python
netflix['date_added'] = netflix['date_added'].astype('string')
netflix['duration'] = netflix['duration'].astype('string')
```

```python
netflix.info()
```

```python
netflix['type'] = netflix['type'].fillna(netflix['type'].mode()[0])
netflix['title'] = netflix['title'].fillna(netflix['title'].mode()[0])
netflix['director'] = netflix['director'].fillna(netflix['director'].mode()[0])
netflix['cast'] = netflix['cast'].fillna(netflix['cast'].mode()[0])
netflix['country'] = netflix['country'].fillna(netflix['country'].mode()[0])
netflix['date_added'] = netflix['date_added'].fillna(netflix['date_added'].mode()[0])
netflix['release_year'] = netflix['release_year'].fillna(netflix['release_year'].mode()[0])
netflix['rating'] = netflix['rating'].fillna(netflix['rating'].mode()[0])
netflix['duration'] = netflix['duration'].fillna(netflix['duration'].mode()[0])
netflix['listed_in'] = netflix['listed_in'].fillna(netflix['listed_in'].mode()[0])
netflix['description'] = netflix['description'].fillna(netflix['description'].mode()[0])
```
